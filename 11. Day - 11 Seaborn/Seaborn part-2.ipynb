# Histogram plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df
```

```python
sns.displot(df["bill_length_mm"])
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240],kde=True) #Kernel density estimator
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240],kde=True, rug=True) #Kernel density estimator
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240],kde=True, rug=True, color="r") 
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240],kde=True, rug=True, color="r", log_scale=True) 
plt.show()
```

## Scatter Plot

```python
df = sns.load_dataset("penguins").head(20)
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex",size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex",size="sex",sizes=(40,100))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex",size="sex",sizes=(60,100)
               ,palette="gist_ncar")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex",size="sex",sizes=(60,100)
               ,palette="gist_ncar",alpha=0.5)
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm", data=df,hue="sex",style="sex",size="sex",sizes=(100,200)
               ,palette="gist_ncar",alpha=1,
               markers=m)
plt.show()
```
