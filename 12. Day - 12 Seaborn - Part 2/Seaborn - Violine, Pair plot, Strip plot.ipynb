# Violin plot in Seaborn

A Violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables sucn that those distributions can be compared.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("tips.csv")
```

```python
df
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, linewidth = 5, palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df) # saturation=0.2
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, color="r") 
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill",linewidth=4, data=df, color="r") 
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill",data=df, hue="sex", split=True)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill",data=df, hue="sex", split=True, scale="width")
plt.show()
```

```python
sns.violinplot(y="day", x="total_bill",data=df, hue="sex")
plt.show()
```

```python
sns.violinplot(x = "time", y="total_bill", data=df, order=["Dinner","Lunch"], inner = "quart") # By default inner = box{"quartile","point","stick",None}
plt.show()
```

```python
sns.violinplot(x=df["total_bill"])
plt.show()
```

```python
sns.violinplot(y=df["total_bill"])
plt.show()
```

# Pair plot

- Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters. 

```python
df1 = pd.read_csv("tips.csv")  #pair plots forms of numerical values
sns.pairplot(df1)
plt.show()
```

```python
sns.pairplot(df1, hue="sex") # Mean curve is formed insted of histogram
plt.show()
```

```python
sns.pairplot(df1, vars=["total_bill", "tip"], hue="sex")
plt.show()
```

```python
sns.pairplot(df1, vars=["total_bill", "tip"], hue="sex", hue_order=["Male","Female"])
plt.show()
```

```python
sns.pairplot(df1, vars=["total_bill", "tip"], hue="sex", hue_order=["Male","Female"], palette='BuGn')
plt.show()
```

```python
sns.pairplot(df1, hue="sex", hue_order=["Male","Female"], x_vars=["tip","total_bill"])
plt.show()
```

```python
sns.pairplot(df1, hue="sex", hue_order=["Male","Female"], y_vars=["tip","total_bill"])
plt.show()
```

```python
sns.pairplot(df1, hue="sex", hue_order=["Male","Female"],kind="reg") #'scatter','kde','hist','reg'
plt.show()
```

```python
sns.pairplot(df1, hue="sex", hue_order=["Male","Female"],kind="reg", diag_kind="hist") #'scatter','kde','hist','reg'
plt.show()
```

```python
sns.pairplot(df1, hue="sex", hue_order=["Male","Female"],markers=["*","<"]) #'scatter','kde','hist','reg'
plt.show()
```

```python
df2= pd.read_csv("tips.csv").head(20)
sns.pairplot(df2, hue="sex", hue_order=["Male","Female"],markers=["*","<"]) #'scatter','kde','hist','reg'
plt.show()
```

# Strip Plot 

- The seaborn. stripplot() method is used to draw a scatter plot where one of variables passed as argument is a categorical variable.

```python
df3= pd.read_csv("tips.csv")
```

```python
df3
```

```python
sns.stripplot(x="day",y="total_bill", data=df3, hue="sex")
plt.plot()
```

```python
sns.stripplot(x="day",y="total_bill", data=df3, hue="sex",palette="rocket_r",linewidth=1, edgecolor="r")
plt.plot()
```

```python
sns.stripplot(x="day",y="total_bill", data=df3, hue="sex",palette="rocket_r",linewidth=1, edgecolor="r",jitter=1)
plt.plot()
```

```python
sns.stripplot(x="day",y="total_bill", data=df3, hue="sex",palette="rocket_r",linewidth=1, edgecolor="r",
             size=10,marker="<")
plt.plot()
```

```python

sns.stripplot(x="day",y="total_bill", data=df3, hue="sex",palette="rocket_r",linewidth=1, edgecolor="r",
             size=10,marker="D",alpha=0.4)
plt.plot()
```

```python
sns.stripplot(x=df3["total_bill"])
plt.plot()
```

# Box Plot

```python
df4 = pd.read_csv("tips.csv")
```

```python
df4
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4)
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4, hue="time")
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4, hue="sex")
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4, hue="sex", color="g")
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4, hue="sex", color="g",order=["Fri","Sun","Thur","Sat"])
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=df4, color="r",order=["Fri","Sun","Thur","Sat"],showmeans=True)
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",
            data=df4, color="w",
            order=["Fri","Sun","Thur","Sat"],
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"})
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",
            data=df4, color="w",
            order=["Fri","Sun","Thur","Sat"],
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},
           linewidth=5)
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",
            data=df4, color="w",
            order=["Fri","Sun","Thur","Sat"],
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},
           linewidth=5,
           palette="plasma")
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",
            data=df4, color="w",
            order=["Fri","Sun","Thur","Sat"],
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},
           linewidth=5,
           palette="plasma",
           orient="v") #Horizontal orientation requires numeric `x` variable.
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="total_bill",y="day",
            data=df4, color="w",
            order=["Fri","Sun","Thur","Sat"],
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},
           linewidth=5,
           palette="plasma",
           orient="h") #Horizontal orientation requires numeric `x` variable.
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(
            data=df4, 
            showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"b"},
           linewidth=5,
           palette="plasma",
           orient="h") #Horizontal orientation requires numeric `x` variable.
plt.plot()
```

```python
sns.set(style="whitegrid")

sns.boxplot(
            x=df4["total_bill"]) 
           
plt.plot()
```
