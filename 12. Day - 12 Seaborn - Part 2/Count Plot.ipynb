# Count Plot

Difference between countplot and barplot??

1. Countplot plots the count of the number of records by category.
2. Barplot plots a value or metric for each category (by default, barplot plots the mean of a variable, by category).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
sns.countplot(x="Sex", data=df)

plt.show()
```

```python
sns.barplot(x="Sex",y="Age" ,data=df)

plt.show()
```

```python
sns.barplot(x="Sex",y="Age" ,data=df, hue="Survived")



plt.show()
```

```python
sns.countplot(x="Sex", data=df,hue="Survived")

plt.show()
```

```python
sns.countplot(y="Sex", data=df,hue="Survived")

plt.show()
```

```python
sns.countplot(x="Sex", data=df,hue="Survived", palette="autumn")

plt.show()
```

```python
sns.countplot(x="Sex", data=df,hue="Survived", color="green")

plt.show()
```

```python
sns.countplot(x="Sex", data=df,hue="Survived", color="green", saturation=0.5)

plt.show()
```
