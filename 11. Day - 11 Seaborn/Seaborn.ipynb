# Seaborn VS Matplotlib

- Seaborn is a data visualization library built on top of matplotlib and closely integrated with pandas data structures in Python.

1. Seaborn offers more appealing default styles and color palettes.
2. Seaborn excels at creating statistical plots.
3. Seaborn works seamlessly with Pandas DataFrames for easy plotting.
4. Seaborn's higher-level interface requires less code for complex plots.
5.  Seaborn includes functions for summary stats and correlations.

## Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,4,1,7,9,7,5]
```

```python
import pandas as pd

x_1 = pd.DataFrame({"var":var, "var_1":var_1})
sns.lineplot(x="var", y="var_1",data=x_1) # It needs dataframe
```

```python
plt.plot(var,var_1)
plt.show() # Using plt.show() , we can remove from printing "[<matplotlib.lines.Line2D at 0x118e90ba410>]" above graph

# We can see y axes is labeled with var_1 and x with var using seaborn
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,4,1,7,9,7,5]

sns.lineplot(x=var, y=var_1) ## Without using DataFrame
plt.show()
```

## Bar plot

```python
var = sns.load_dataset("penguins")
```

```python
var
```

```python
sns.barplot(x = var.island, y = var.bill_depth_mm)
plt.show() # matplotlib function
```

```python
sns.barplot(x = "island", y = "bill_depth_mm",data=var)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex",order=order_1)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex",order=order_1, hue_order=["Female","Male"])
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex",order=order_1, hue_order=["Female","Male"], errorbar=('ci', 100))
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex",order=order_1, hue_order=["Female","Male"], errorbar=('ci', 100), n_boot=2) #maximum value of ci is 100 only.  
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",data=var,hue="sex",order=order_1, hue_order=["Female","Male"], errorbar=('ci', 100), orient = "v") ## Horizontal orientation requires numeric `x` variable, categorical value is not allowed.
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "bill_depth_mm", y = "bill_depth_mm",data=var, orient = "v") ## Horizontal orientation requires numeric `x` variable, categorical value is not allowed.
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm",data=var, color="green")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_depth_mm",data=var, color="green",palette="winter")
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=0.1)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=10, errcolor ='y', errwidth=10)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=10, errcolor ='b',capsize=0.5)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=100, errcolor ='b',dodge = False)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=100, errcolor ='b', alpha=1)
plt.show()
```

```python
##sns.set(style="darkgrid")
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=100, errcolor ='b', alpha=1)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_depth_mm",
            data=var,hue="sex",order=order_1, 
            hue_order=["Female","Male"], orient = "v",
           saturation=100, errcolor ='b')
plt.show()
```
